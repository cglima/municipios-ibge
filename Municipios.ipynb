```python
import pandas as pd
```

```python
cidades = pd.read_json('https://servicodados.ibge.gov.br/api/v1/localidades/municipios')
cidades
```

```python
micro = pd.json_normalize(data=cidades.microrregiao).rename(columns={
    'id': 'microrregiao_id', 
    'nome': 'microrregiao_nome', 
    'mesorregiao.id': 'mesorregiao_id',
    'mesorregiao.nome': 'mesorregiao_nome',
    'mesorregiao.UF.id': 'uf_id',
    'mesorregiao.UF.nome': 'uf_nome',
    'mesorregiao.UF.sigla': 'uf_sigla',
    'mesorregiao.UF.regiao.id': 'regiao_id',
    'mesorregiao.UF.regiao.sigla': 'regiao_sigla',
    'mesorregiao.UF.regiao.nome': 'regiao_nome',    
})
micro
```

```python
nome_id_municipio = cidades[['id', 'nome']].rename(columns={'id': 'municipio_id', 'nome': 'municipio_nome'})
nome_id_municipio
```

```python
municipios = nome_id_municipio.join(micro)
municipios
```

```python
microrregioes = municipios.groupby(by="microrregiao_nome")
microrregioes = microrregioes.first()
microrregioes
```

```python
type(microrregioes)
```

```python
microrregioes.reset_index(inplace=True)
microrregioes
```

```python
microrregioes = microrregioes[['municipio_id', 'municipio_nome',
                               'microrregiao_id', 'microrregiao_nome',
                               'mesorregiao_id', 'mesorregiao_nome', 
                              'uf_id', 'uf_sigla', 'uf_nome',
                               'regiao_id', 'regiao_sigla', 'regiao_nome']]
microrregioes
```

```python
microrregioes[['municipio_id', 'municipio_nome']] = None
microrregioes
```

```python
mesorregioes = microrregioes.groupby('mesorregiao_nome')
mesorregioes = mesorregioes.first()
mesorregioes.reset_index(inplace=True)
mesorregioes
```

```python
mesorregioes = mesorregioes[['municipio_id', 'municipio_nome',
                               'microrregiao_id', 'microrregiao_nome',
                               'mesorregiao_id', 'mesorregiao_nome', 
                              'uf_id', 'uf_sigla', 'uf_nome',
                               'regiao_id', 'regiao_sigla', 'regiao_nome']]
mesorregioes
```

```python
mesorregioes[['microrregiao_id', 'microrregiao_nome']] = None
mesorregioes
```

```python
estados = mesorregioes.groupby('uf_nome')
estados = estados.first()
estados.reset_index(inplace=True)
estados
```

```python
estados = estados[['municipio_id', 'municipio_nome',
                               'microrregiao_id', 'microrregiao_nome',
                               'mesorregiao_id', 'mesorregiao_nome', 
                              'uf_id', 'uf_sigla', 'uf_nome',
                               'regiao_id', 'regiao_sigla', 'regiao_nome']]
estados
```

```python
estados[['mesorregiao_id', 'mesorregiao_nome']] = None
estados
```

```python
regioes = estados.groupby('regiao_nome')
regioes = regioes.first()
regioes.reset_index(inplace=True)
regioes
```

```python
regioes = regioes[['municipio_id', 'municipio_nome',
                               'microrregiao_id', 'microrregiao_nome',
                               'mesorregiao_id', 'mesorregiao_nome', 
                              'uf_id', 'uf_sigla', 'uf_nome',
                               'regiao_id', 'regiao_sigla', 'regiao_nome']]
regioes
```

```python
regioes[['uf_id', 'uf_nome', 'uf_sigla']] = None
regioes
```

```python
agrupados = pd.concat([municipios, microrregioes, mesorregioes, estados, regioes])
agrupados.reset_index(drop=True)
```

```python
agrupados.to_csv('municipios.csv', sep=';', index=False)
```
